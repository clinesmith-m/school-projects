# Dropping extra symbols.

```python
def fixSqMi(df):
    sq1 = df['Area'].str.extract("([0-9]+)", expand = True)
    sq2 = df['Area'].str.extract("([.])", expand = True)
    sq3 = df['Area'].str.extract("([0-9]+) ", expand = True)
    sqMile = sq1 + sq2 + sq3
    df['Area'] = sqMile
```

```python
def fixHouseholdSize(df):
    occupants = df['Average Household Size'].str.extract("([0-9.]{3})", expand = True)
    df['Average Household Size'] = occupants
```

```python
def fixPop(item):
    popList = item.split(",")
    pop = "".join(popList)
    return pop
```

```python
def fixMedAge(df, colName):
    medAge = df[colName].str.extract("([\d.]{4})", expand = True)
    df[colName] = medAge
```

```python
def fixComma(item):
    if type(item) == float:
        return item
    item = item.replace("$","").replace(",","")
    return item
```

```python
def fixPercentage(df, colName):
    percent = df[colName].str.extract("([0-9.]+)", expand = True)
    df[colName] = percent
```

```python
def fixPopDensity(item):
    if type(item) == float:
        return item
    item = item.replace(",", "")
    itemList = item.split(" ")
    item = itemList[0]
    return item
```

```python
fixSqMi(df)
fixHouseholdSize(df)
fixMedAge(df, "Median Age Among Men")
df['Female Population'] = df['Female Population'].apply(fixComma)
df['Male Population'] = df['Male Population'].apply(fixComma)
fixMedAge(df, "Median Age Among Men")
fixMedAge(df, "Median Age Among Women")
df['Median Income'] = df['Median Income'].apply(fixComma)
df['Median Rent'] = df['Median Rent'].apply(fixComma)
fixPercentage(df, "Percentage of Family Households")
fixPercentage(df, "Percentage of Married Families")
fixPercentage(df, "Percentage of Married Families with Children")
fixPercentage(df, "Percentage of Never-Married Men")
fixPercentage(df, "Percentage of Never-Married Women")
fixPercentage(df, "Percentage of People Who Speak English Poorly or Not at All")
fixPercentage(df, "Percentage of Single Mother Families")
df["Population"] = df["Population"].apply(fixComma)
df["Population Density"] = df["Population Density"].apply(fixPopDensity)
fixPercentage(df, "Residents Born Foreign")
fixPercentage(df, "Residents Born as U.S. Citizens Outside the U.S.")
fixPercentage(df, "Residents Born in Texas")
fixPercentage(df, "Residents Born in the U.S.")
```

# Making an Address Column

```python
def fixNames(df):
    expandedNames = df['Neighborhood Name'].str.extract("([\w /]+)", expand = True)
    expandedNames = expandedNames + ", Fort Worth, TX"
    return expandedNames

df["Address"] = fixNames(df)
df.head()
```

# Getting Latitude and Longitude

```python
from geopy.geocoders import Nominatim
```

```python
def makeAddrDict(item):
    hood = item
    city = "Fort Worth"
    state = "TX"
    addrDict = dict(zip(["street", "city", "state"], [hood, city, state]))
    return addrDict

def getNeighCoords(inputAddress):
    inputAddress = makeAddrDict(inputAddress)
    try:
        geolocator = Nominatim(user_agent = 'my-application')
        location = geolocator.geocode(inputAddress)
        return location.latitude,location.longitude
    except:
        return None,None
```

```python
df["Coordinates"] = df["Neighborhood Name"].apply(lambda x:getNeighCoords(x))
df["Latitude"] = df["Coordinates"].apply(lambda x:x[0])
df["Longitude"] = df["Coordinates"].apply(lambda x:x[1])
```

# Getting Venue Data

```python
with open("foursquareID.txt", "r") as f:
    CLIENT_ID = f.readline().rstrip()
    CLIENT_SECRET = f.readline().rstrip()

VERSION = "20190401"
```

```python
import requests 
def getNearbyVenues(coords, radius = 500, limit = 100):
    # names, latitudes, longitudes = zip(*coords)
    venues_list=[]
    for c in coords:
        name, lat, lon = c['Neighborhood Name'],c['Latitude'],c['Longitude']
        try:
            # create the API request URL
            url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
                CLIENT_ID, 
                CLIENT_SECRET, 
                VERSION, 
                lat, 
                lon, 
                radius, 
                limit)
            #print(url)
            # make the GET request
            results = requests.get(url).json()["response"]['groups'][0]['items']
            #print(results)

            # return only relevant information for each nearby venue
            venues_list.append([(
                name, 
                v['venue']['name'], 
                v['venue']['categories'][0]['name']) for v in results])
        except:
            pass
    
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood Name', 
                  'Venue', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
df_venues = getNearbyVenues(df[['Neighborhood Name','Latitude','Longitude']].dropna().to_dict('records'))
df_venues.head()
```

```python
df_dummies = pd.get_dummies(df_venues["Venue Category"])
df_dummies.head()
```

```python
df_dummies["Neighborhood Name"] = df_venues["Neighborhood Name"]
df_dummies.head()
```

```python
df_dummies = df_dummies.groupby("Neighborhood Name").mean()
df_dummies.head()
```

```python
colNames = list(df_dummies.sum().sort_values(ascending = False).index[:50])
```

```python
df_dummies_trimmed = df_dummies[colNames]
df_dummies_trimmed.head()
```

```python
df_with_venues = df.join(df_dummies_trimmed)
```

```python
# Cleaning empty data.
findNone(df_with_venues, "Coordinates")
findNone(df_with_venues, "Latitude")
findNone(df_with_venues, "Longitude")
for colName in colNames:
    findNone(df_with_venues, colName)

df_with_venues.head()
```
